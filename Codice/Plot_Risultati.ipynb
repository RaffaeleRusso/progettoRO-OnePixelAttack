# Import librerie

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

# Plot

```python
X = ['Popolazione 100','Popolazione 400','Popolazione 1000']
pixels1 = [0.93, 0.93, 0.93]
pixels10 = [0.67, 0.76, 0.75]
pixels20 = [0.53, 0.64, 0.63]
pixels30 = [0.49, 0.54, 0.42]
pixels50 = [0.38, 0.37, 0.36]
  
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, pixels1, 0.15, label = 'pixels1')
plt.bar(X_axis - 0.05, pixels10, 0.15, label = 'pixels10')
plt.bar(X_axis + 0.1, pixels20, 0.15, label = 'pixels20')
plt.bar(X_axis + 0.25, pixels30, 0.15, label = 'pixels30')
plt.bar(X_axis + 0.4, pixels50, 0.15, label = 'pixels50')

plt.xticks(X_axis, X)
plt.ylabel("Accuracy")
plt.axhline(y = 0.99, color = 'turquoise', label = "Acc Pre_Att")
plt.legend(loc = 'upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0)
plt.title("CR = 1, F = (0.5, 1), Iterazioni = 50")

#plt.savefig('Andamento_pixels_it50.png')
plt.show()
```
